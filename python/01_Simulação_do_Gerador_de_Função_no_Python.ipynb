# Etapa 2: Setup da Simulação

```python
GridFreq = 100  #Nominal Grid Frequency in Hz
Fs       = 10000 #Sampling frequency = 10Khz
Tfinal   = 0.1  #Time the simulation

# Variaveis calculadas
T        = 1/GridFreq # Periodo do Sinal
Ts       = 1/Fs #Sampling Time = 1/Fs
time     = arange(0,Tfinal,Ts) #Simulation Time vector
w        = 2*pi*GridFreq #Frequencia Angular
```

# Etapa 3: Geração de uma onda dente de Serra

```python
# Gera pontos como se estivesse lendo tensões alternadas.
A = 10
onda = 0
denteSerra = array([])
for n in time:
  onda=(A / T)*(n % T) # y = m*x onde m = A/T e x = n % T
  denteSerra =  append(denteSerra, onda)

fig, ax = subplots(figsize=(20,6), facecolor='#DEDEDE')
title("Calibração|Ajuste");
line1 = ax.plot(time,denteSerra,'b') #linewidth=3
#legend([line1],["Dente de Serra"]);
xlabel('Tempo')
ylabel('Amplitude')
```

# Etapa 4: Simulação de uma onda quadrada.

```python
A = 10
onda = 0
dutyCicle = 0.6
quadrada = array([])
for n in time:
  if n % T <= dutyCicle * T :
    onda = A
  else:
    onda = 0
  quadrada = append(quadrada, onda)

fig, ax = subplots(figsize=(25,6), facecolor='#DEDEDE')
title("Calibração|Ajuste");
line1 = ax.plot(time, quadrada, 'b', linewidth=3, label="Onda Quadrada")
ax.legend(line1,"Onda Quadrada");
xlabel('Tempo')
ylabel('Amplitude')
```

# Etapa 5: Simulação de um *onda* Triangular.

```python
A = 5
onda = 0
triang = array([])
for n in time:
  onda = (A/(T/2))*(n % T)  # y = m*x onde m = A/(T/2) e x = n % T
  if (n % T) <= T / 2 :
    triang = append(triang, onda)
  else:
     triang = append(triang, 2*A-onda)


fig, ax = subplots(figsize=(20,6), facecolor='#DEDEDE')
title("Calibração|Ajuste");
line1 = ax.plot(time,triang,'b') #linewidth=3
legend(line1,"Onda Triangular");
xlabel('Tempo')
ylabel('Amplitude')
```

# Etapa 6: Simulação de um *onda* Senoidal.

```python
T = 20
t = arange(0,100,0.1)
y = sin(2*pi*(1/T)*t)
fig, ax = subplots(figsize=(20,6), facecolor='#DEDEDE')
title("Aula 01 de Instrumentação Industrial")
line1 = ax.plot(t,y,'b')
legend(line1,"SIN")
```
